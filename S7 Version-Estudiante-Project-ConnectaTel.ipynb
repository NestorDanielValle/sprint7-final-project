## 🧩Paso 4: Summary statistics de uso por usuario


### 4.1 Agrupación por comportamiento de uso

🎯**Objetivo**: Resumir las variables clave de la tabla `usage` **por usuario**, creando métricas que representen su comportamiento real de uso histórico. 

**Instrucciones:**: 
1. Construye una tabla agregada de `usage` por `user_id` que incluya:
- número total de mensajes  
- número total de llamadas  
- total de minutos de llamadas

2. Renombra las columnas para que tengan nombres claros:  
- `cant_mensajes`  
- `cant_llamadas`  
- `cant_minutos_llamada`
3. Combina esta tabla con `users`.

```python
# Columnas auxiliares
usage["is_text"] = (usage["type"] == "text").astype(int) #conocer el total de mensajes
usage["is_call"] = (usage["type"] == "call").astype(int) #conocer el total de llamadas


# Agrupar información por usuario
usage_agg = (usage.groupby("user_id").agg(
    texto_suma = ("is_text", "sum"), 
    llamada_suma = ("is_call", "sum"),
    duracion_suma = ("duration", "sum"))
    ).reset_index()
# observar resultado
usage_agg.head(3)
```

```python

# Renombrar columnas
usage_agg.rename(columns={
    'texto_suma': 'cant_mensajes', 
    'llamada_suma': 'cant_llamadas', 
    'duracion_suma': 'cant_minutos_llamada'},
                 inplace=True)
# observar resultado
usage_agg.head(3)
```

```python
# Combinar la tabla agregada con el dataset de usuarios
user_profile = users.merge(usage_agg, on="user_id", how="left")
user_profile.head(5)
```

- Después de crear los gráfico, responde si **existen o no outliers** en las variables.  
- Si hay outliers, crea otro bucle para calcular los límites de esas columnas usando el **método IQR** y decide qué hacer con ellos.

- Dentro del bucle, usa `plt.title(f'Boxplot: {col}')` para que el título cambie acorde a la columna.

```python
# Visualizando usando BoxPlot 
columnas_numericas = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

for col in columnas_numericas:
    plt.figure(figsize=(8, 4))

    sns.boxplot(x=user_profile[col])

    plt.title(f'Boxplot: {col}')
    plt.xlabel(col)
    plt.show()
```

💡Insights: 
- Age: Presenta pocos o ningún outlier. Las edades se concentran dentro de un rango esperado, por lo que no parece haber valores extremos que requieran tratamiento adicional.
- cant_mensajes: Presenta outliers en el límite superior. Existen usuarios que envían una cantidad de mensajes significativamente mayor que el resto, lo que puede corresponder a clientes con un uso intensivo del servicio.
- cant_llamadas: Presenta outliers en el límite superior. Algunos usuarios realizan muchas más llamadas que el promedio, reflejando patrones de consumo elevados.
- cant_minutos_llamada: Presenta outliers en el límite superior. Se observan clientes con un número muy alto de minutos de llamada, lo cual representa un comportamiento válido y útil para identificar usuarios de alto consumo.

```python
# Calcular límites con el método IQR
columnas_limites = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

for col in columnas_limites:
    Q1 = user_profile[col].quantile(0.25)
    Q3 = user_profile[col].quantile(0.75)

    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    print(f'\n{col}')
    print(f'Q1: {Q1}')
    print(f'Q3: {Q3}')
    print(f'IQR: {IQR}')
    print(f'Límite superior: {limite_superior:.2f}')
    print(f'Limite inferior: {limite_inferior:.2f}')
```

```python
# Revisa los limites superiores y el max, para tomar la decisión de mantener los outliers o no
user_profile[columnas_limites].describe()
```

💡Insights: 
- cant_mensajes: mantener o no outliers, porqué? Mantener los outliers. Se identifican usuarios que envían más de 11.5 mensajes, pero estos valores pueden corresponder a clientes con un uso intensivo del servicio. Eliminarlos podría ocultar patrones importantes para la segmentación y el diseño de nuevos planes.
- cant_llamadas: mantener o no outliers, porqué? Mantener los outliers. Existen usuarios que realizan más de 10.5 llamadas, lo que refleja un comportamiento de alto consumo y no necesariamente un error en los datos. Estos casos son relevantes para identificar clientes con necesidades específicas.
- cant_minutos_llamada: mantener o no outliers, porqué? Mantener los outliers. Los usuarios que superan los 61.86 minutos de llamadas representan clientes con un consumo elevado. Estos valores son útiles para analizar patrones de uso, detectar posibles oportunidades comerciales y evaluar si los planes actuales satisfacen las necesidades de este segmento.

## 🧩Paso 6: Segmentación de Clientes

### 6.1 Segmentación de Clientes Por Uso

🎯 **Objetivo:** Clasificar a cada usuario en un grupo de uso (Bajo uso, Uso medio, Alto uso) basándose en la cantidad de llamadas y mensajes registrados.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_uso` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones de llamadas y mensajes y asigna:
  - `'Bajo uso'` cuando llamadas < 5 y mensajes < 5
  - `'Uso medio'` cuando llamadas < 10 y mensajes < 10
  - `'Alto uso'` para el resto de casos

```python
# Crear columna grupo_uso
import numpy as np

condiciones = [
    (user_profile['cant_llamadas'] < 5) & (user_profile['cant_mensajes'] < 5),
    (user_profile['cant_llamadas'] < 10) & (user_profile['cant_mensajes'] < 10)
]

grupos = ['Bajo uso','Uso medio']

user_profile['grupo_uso'] = np.select(
    condiciones,
    grupos,
    default='Alto uso'
)
```

```python
# verificar cambios
user_profile.head()
```

### 6.2 Segmentación de Clientes Por Edad

🎯 **Objetivo:**: Clasificar a cada usuario en un grupo por **edad**.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_edad` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones y asigna:
  - `'Joven'` cuando age < 30
  - `'Adulto'` cuando age < 60
  - `'Adulto Mayor'` para el resto de casos

```python
# Crear columna grupo_edad
import numpy as np

condiciones = [
    (user_profile['age'] < 30) & (user_profile['age'] < 30),
    (user_profile['age'] < 60) & (user_profile['age'] < 60)
]

grupos = ['Joven','Adulto']

user_profile['grupo_edad'] = np.select(
    condiciones,
    grupos,
    default='Adulto_Mayor'
)
```

### Análisis ejecutivo

⚠️ **Problemas detectados en los datos**
- Se identificaron 469 valores nulos (11.73%) y 96 registros con el valor "?" en la columna city, por lo que se reemplazó el sentinel por valores nulos (pd.NA).
La columna age contenía el valor inválido -999, el cual fue reemplazado por la mediana (47 años). Además, se detectaron 40 fechas de registro del año 2026, las cuales fueron marcadas como nulas por estar fuera del periodo de análisis.


🔍 **Segmentos por Edad**
- La mayoría de los usuarios se concentra en edades intermedias (entre 33 y 63 años), con una mediana de 48 años.
No se observaron diferencias importantes en la distribución de edades entre los planes Básico y Premium, por lo que la edad, por sí sola, no parece explicar el tipo de plan contratado.

📊 **Segmentos por Nivel de Uso**
- Se clasificó a los clientes en Bajo uso, Uso medio y Alto uso según la cantidad de llamadas y mensajes.
Se identificaron usuarios con un consumo muy elevado de llamadas, mensajes y minutos, considerados outliers, los cuales se conservaron por representar comportamientos reales de clientes de alto consumo.


➡️ Esto sugiere que existen distintos perfiles de clientes y que aquellos con mayor nivel de uso representan una oportunidad para ofrecer planes con mayores beneficios y estrategias de fidelización.

💡 **Recomendaciones**
- Diseñar planes específicos para clientes de alto consumo, incluyendo mayores beneficios en llamadas y mensajes.
Implementar campañas para migrar usuarios de uso medio hacia planes con mayor capacidad mediante promociones personalizadas.
Fortalecer los procesos de validación de datos para evitar valores inválidos, sentinels y fechas fuera de rango en futuras bases de datos.
Dar seguimiento a los clientes de mayor consumo para mejorar su experiencia y aumentar la retención mediante ofertas adaptadas a sus necesidades.
